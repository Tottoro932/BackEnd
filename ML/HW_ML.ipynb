### Предварительная обработка данных

1) Job_Placement_Data.csv

```python
# удалим колонки ssc_board и hsc_board - нам не важно, каким был "Совет по образованию"

job_data.drop(['ssc_board','hsc_board'],axis='columns', inplace=True)
job_data.head(n=10)
```

```python
# проверим наличие пропусков в данных
job_data.isnull().sum()
```

```python
# приведем категориальные данные к числовому типу

# закодируем признак, который будем предсказывать (был опыт работы или нет)
map_dic = {
    'Yes':0,
    'No':1
}
job_data['work_experience'] = job_data['work_experience'].map(map_dic)
job_data.head(n=10)
```

```python
# для всех оставшихся признаков применим one-hot кодирование
cat_features = ['gender', 'hsc_subject', 'undergrad_degree', 'specialisation', 'status']
prefixes = ['g', 'hs', 'ud', 'sp', 'st']
```

```python
job_data = pd.get_dummies(columns=cat_features, data=job_data, prefix=prefixes)
job_data.head(n=10)
```

```python
# сохраняем предобработанный датасет
job_data.to_csv('C:/PRJ/PYTHON/BackEnd/ML/Job_Placement_Data_preprocessed.csv', index=False)
```

2) Steel_industry_data.csv

### Решение задачи классификации
Задача: предсказать был у человека опыт работы (0) или его не было(1)

```python
# загрузим данные
data_classification = pd.read_csv("C:/PRJ/PYTHON/BackEnd/ML/Job_Placement_Data_preprocessed.csv")
```

```python
# выделим целевой признак
y_classification = data_classification['work_experience']

# выделим все, кроме целевого признака - по этим значениям будем предсказывать целевой признак
X_classification = data_classification.drop(columns = ['work_experience'])
```

```python
from sklearn.model_selection import train_test_split
# разобьем данные регрессии
X_classification_train, X_classification_test, y_classification_train, y_classification_test = train_test_split(X_classification,
                                                                                                                y_classification,
                                                                                                                stratify=y_classification,
                                                                                                                test_size=0.2)
```

```python
len(X_classification_train), len(X_classification_test)
```

```python
# здесь импортируем сами модели
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, GradientBoostingClassifier

# это функции, которые помогут оценивать качество моделей
from sklearn.metrics import confusion_matrix, classification_report

# импортируем класс GridSearchCV для подбора гиперпараметра alfa
from sklearn.model_selection import GridSearchCV
```

1) K ближайших соседей

```python
parameters_k = {'n_neighbors': [1, 3, 5, 7, 9, 11, 13, 15]}

k_optimal = GridSearchCV(KNeighborsClassifier(), parameters_k).fit(X_classification_train, y_classification_train)
k_optimal.best_params_
```

```python
k_optimal.predict([X_classification_test.iloc[0]])
```

```python
print(classification_report(y_classification_test, k_optimal.predict(X_classification_test)))
```

2) Логистическая регрессия

```python
parameters_l = {'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']}

l_optimal = GridSearchCV(LogisticRegression(), parameters_l).fit(X_classification_train, y_classification_train)
l_optimal.best_params_
```

```python
print(classification_report(y_classification_test, l_optimal.predict(X_classification_test)))
```

3) Деревья принятия решений

```python
parameters_dr = {'criterion': ['gini', 'entropy', 'log_loss']}

dr_optimal = GridSearchCV(DecisionTreeClassifier(), parameters_dr).fit(X_classification_train, y_classification_train)
dr_optimal.best_params_
```

```python
print(classification_report(y_classification_test, dr_optimal.predict(X_classification_test)))
```

### Глубокая нейронная сеть для задачи классификации

```python
import tensorflow as tf
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# данные по классам не сбалансированы, поэтому:

w0 = 1 / y_classification_train[y_classification_train==0].shape[0]
w1 = 1 / y_classification_train[y_classification_train==1].shape[0]
```

```python
model_classification = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(256, activation="relu", input_shape=(17,)),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dropout(0.05),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(32, activation="relu"),
        tf.keras.layers.Dropout(0.05),
        tf.keras.layers.Dense(16, activation="relu"),
        # сначала используем 1 нейрон и sigmoid
        tf.keras.layers.Dense(1, activation="sigmoid"),
    ]
)
```

```python
model_classification.summary()
```

```python
model_classification.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="binary_crossentropy")
model_classification.fit(X_classification_train, y_classification_train, epochs=25, verbose=None,
                           class_weight={0: w0, 1: w1})
y_pred = np.around(model_classification.predict(X_classification_test))
```

```python
print(classification_report(y_classification_test, y_pred))
print(confusion_matrix(y_classification_test, y_pred))
```

### Глубокая нейронная сеть для задачи регрессии

```python
model_regression = tf.keras.Sequential(
    [
        # Dense - полносвязный слой (каждый нейрон следующего слоя связан со всеми нейронами предыдущего)
        tf.keras.layers.Dense(256, activation="elu", input_shape=(32,11)),
        # на втором скрытом слое будет 32 нейрона
        tf.keras.layers.Dense(128, activation="elu"),
        # Dropout позволяет внести фактор случайности - при обучении часть нейронов будет отключаться
        # каждый нейрон, в данном случае, будет отключаться с вероятностью 0.15
        tf.keras.layers.Dropout(0.15),
        tf.keras.layers.Dense(32, activation="elu"),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(8, activation="elu"),
        tf.keras.layers.Dropout(0.15),
        # на выходе один нейрон
        tf.keras.layers.Dense(1, activation="linear"),
    ]
)
```

```python
model_regression.summary()
```

```python
# компелируем
model_regression.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="mse")

# обучение
model_regression.fit(X_regression_train, y_regression_train, epochs=10, verbose=None)
```

```python
print(mean_absolute_error(y_regression_test, model_regression.predict(X_regression_test)))
print(mean_squared_error(y_regression_test, model_regression.predict(X_regression_test)))
```
